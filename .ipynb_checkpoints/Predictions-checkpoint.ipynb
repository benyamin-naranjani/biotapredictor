```python
import pandas as pd

# Specify the file path
file_path_t = 'test_canonical.xlsx'

# Load the Excel file into a Pandas DataFrame
df_t = pd.read_excel(file_path_t)

# Display the DataFrame
from rdkit import Chem
import pandas as pd

# Sample DataFrame creation
# df = pd.DataFrame({'Unique Isomeric Smiles': ['C[C@@H](N)C(=O)O', 'CC(N)C(=O)O', ...]})

# Function to convert isomeric SMILES to canonical SMILES without stereochemistry
def to_canonical(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        # Remove stereochemistry information
        Chem.RemoveStereochemistry(mol)
        return Chem.MolToSmiles(mol, isomericSmiles=False)
    else:
        return None

# Apply the function to the 'Unique Isomeric Smiles' column
df_t['Canonical Smiles'] = df_t['Isomeric SMILES'].apply(to_canonical)

# Remove duplicates based on the 'Canonical Smiles' column
df_unique_t = df_t.drop_duplicates(subset='Canonical Smiles')

# Optionally, you can drop the original 'Unique Isomeric Smiles' column if no longer needed
# df_unique = df_unique.drop(columns=['Unique Isomeric Smiles'])

# Reset index if desired
df_unique_t.reset_index(drop=True, inplace=True)

df_t = df_unique_t
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
import pandas as pd

# Assuming df_l is your DataFrame and it has been loaded previously
# df_l = pd.read_csv('your_dataset.csv')  # Example of loading the DataFrame if needed

# Initialize the SDF writer to save all best conformers
writer = Chem.SDWriter('test_dataset.sdf')

# Iterate over each SMILES string in the "Canonical Smiles" column of df_l
for smiles in df_t["Canonical Smiles"]:
    molecule = Chem.MolFromSmiles(smiles)
    molecule = Chem.AddHs(molecule)
    AllChem.EmbedMultipleConfs(molecule, numConfs=50, useExpTorsionAnglePrefs=True, useBasicKnowledge=True)

    lowest_energy = float('inf')
    lowest_energy_confId = None

    for confId in range(molecule.GetNumConformers()):
        if AllChem.MMFFHasAllMoleculeParams(molecule):
            mmff_props = AllChem.MMFFGetMoleculeForceField(molecule, AllChem.MMFFGetMoleculeProperties(molecule), confId=confId)
            energy = mmff_props.CalcEnergy()
        else:
            uff_props = AllChem.UFFGetMoleculeForceField(molecule, confId=confId)
            energy = uff_props.CalcEnergy()

        if energy < lowest_energy:
            lowest_energy = energy
            lowest_energy_confId = confId

    # Save only the conformer with the lowest energy
    if lowest_energy_confId is not None:
        writer.write(molecule, confId=lowest_energy_confId)

writer.close()

print(f"All best conformers saved to test_dataset.sdf'.")
```
